# ANN Based Classification of Breast Cancer

## **Import Libraries**

```python
#Import Libraries
import numpy as np
import pandas as pd
import keras
from keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

##**Load the Dataset**

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
```

```python
data = pd.read_csv(url, header=None)
data
```

```python
column_names = ['id', 'diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
data.columns = column_names
```

```python
print(data.columns)
```

```python
print(url)
```

##**Preprocessing**

```python
data.isnull().sum()
```

```python
data.dtypes
```

##Normalize Labels


```python


label_encoder = LabelEncoder()

data['diagnosis'] = label_encoder.fit_transform(data['diagnosis'])
data.head()
```

```python
Malignant,Benign = data['diagnosis'].value_counts()
print('Benign',Benign)
print('Malignant',Malignant)
```

## Splitting the dataset into features and target

```python
x = data.drop(['id', 'diagnosis'], axis=1)
y = data['diagnosis']
```

## Feature Scaling

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

## Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

## Building the Neural Network

```python
from keras.models import Sequential

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

model = Sequential([
    Dense(16, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid'),

])
```

```python
model.summary()
```

## Compiling the model

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


```

## Training the model

```python
model.fit(x_train, y_train, batch_size=150, epochs=500)
```

## Accuracy

```python
accuracy = model.evaluate(x_test, y_test)
print('Accuracy on test set:', accuracy)
```

Accuracy is 96.49% which indicates that the model is able to classify the majority of the samples in test set.

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
```

```python
print('Test accuracy:', round(test_acc,4))
```

## Confusion Matrix

```python
# Predicting probabilities on the test set
y_pred_prob = model.predict(x_test)
# Thresholding the probabilities to get the predicted classes
y_pred = (y_pred_prob > 0.5).astype(int)

# Generating the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
